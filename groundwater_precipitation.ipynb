```python
import glob
import pandas as pd
import os

# Define the path to the rainfall data directory
rainfall_data_path = '/content/rainfall_data'

# Find all CSV files in the rainfall data directory
rainfall_files = sorted(glob.glob(os.path.join(rainfall_data_path, "*.csv")))

# Check if any rainfall files were found
if not rainfall_files:
    print(f"No CSV files found in {rainfall_data_path}")
else:
    # Read each rainfall CSV file into a DataFrame and store in a list
    rainfall_df_list = []
    for f in rainfall_files:
        try:
            df = pd.read_csv(f)
            rainfall_df_list.append(df)
        except Exception as e:
            print(f"Error reading {f}: {e}")

    # Check if any DataFrames were successfully read
    if rainfall_df_list:
        # Concatenate all rainfall DataFrames in the list into a single DataFrame
        rainfall_data = pd.concat(rainfall_df_list, ignore_index=True)

        # Remove duplicate rows
        initial_rainfall_rows = len(rainfall_data)
        rainfall_data = rainfall_data.drop_duplicates()
        rows_after_rainfall_duplicates = len(rainfall_data)

        print(f"Rainfall data loaded and concatenated.")
        print(f"Rows before dropping duplicates: {initial_rainfall_rows}")
        print(f"Rows after dropping duplicates: {rows_after_rainfall_duplicates}")
        print(f"Duplicate rows removed: {initial_rainfall_rows - rows_after_rainfall_duplicates}")

        # Display the first few rows of the processed rainfall data
        print("\nProcessed Rainfall Data Head (Duplicates Removed):")
        display(rainfall_data.head())

    else:
        print("No rainfall DataFrames were successfully read.")
```

```python
import pandas as pd
import numpy as np

# Step 1: Use the actual pre-processed DataFrames
# groundwater_data and rainfall_data are already available from previous steps.

# Step 2: Generate time series using the actual number of original files
# We need the number of original files from the loading steps.
# Assuming 'groundwater_files' and 'rainfall_files' variables are available and accurate.
if 'groundwater_files' in globals() and groundwater_files and 'rainfall_files' in globals() and rainfall_files:
    num_groundwater_files = len(groundwater_files)
    num_rainfall_files = len(rainfall_files)

    # Ensure the number of files is consistent or handle if they differ in time steps
    if num_groundwater_files != num_rainfall_files:
        print(f"Warning: Number of groundwater files ({num_groundwater_files}) and rainfall files ({num_rainfall_files}) are different.")
        print("Proceeding with the smaller number of files to align time periods, but this might require further investigation.")
        num_files = min(num_groundwater_files, num_rainfall_files)
    else:
        num_files = num_groundwater_files # Or num_rainfall_files, they are the same

    if len(groundwater_data) > 0 and len(rainfall_data) > 0 and num_files > 0:
        # Assume data is ordered chronologically by original file concatenation order.
        # Divide DataFrames into chunks based on the number of original files.
        rows_per_groundwater_chunk = len(groundwater_data) // num_files
        rows_per_rainfall_chunk = len(rainfall_data) // num_files

        print(f"\nAggregating into {num_files} time periods.")

        # Groundwater anomaly time series: mean 'lwe' per chunk
        # Ensure 'lwe' column exists
        if 'lwe' in groundwater_data.columns:
            groundwater_time_series = [
                groundwater_data.iloc[i * rows_per_groundwater_chunk : (i + 1) * rows_per_groundwater_chunk]['lwe'].mean()
                for i in range(num_files)
            ]
            print("\nGroundwater Time Series (Anomaly in m) created.")
        else:
            print("Error: 'lwe' column not found in groundwater_data. Cannot create groundwater time series.")
            groundwater_time_series = []


        # Rainfall time series: mean 'Precipitation(mm/hr)' per chunk
        # Ensure the chosen precipitation column exists. We decided to keep 'Precipitation(mm/hr)'.
        precipitation_column = 'Precipitation(mm/hr)'
        if precipitation_column in rainfall_data.columns:
            rainfall_time_series = [
                rainfall_data.iloc[i * rows_per_rainfall_chunk : (i + 1) * rows_per_rainfall_chunk][precipitation_column].mean()
                for i in range(num_files)
            ]
            print(f"Rainfall Time Series ({precipitation_column}) created.")
        else:
            print(f"Error: Precipitation column '{precipitation_column}' not found in rainfall_data. Cannot create rainfall time series.")
            rainfall_time_series = []

        # Step 3: Display as tidy DataFrames for easy viewing/export
        if groundwater_time_series:
            gw_ts_df = pd.DataFrame({
                'Period': range(1, num_files + 1),
                'Groundwater Anomaly (m)': groundwater_time_series
            })
            print("\nGroundwater Time Series DataFrame:")
            display(gw_ts_df.head())
        else:
            print("\nGroundwater time series DataFrame not created.")

        if rainfall_time_series:
             rain_ts_df = pd.DataFrame({
                'Period': range(1, num_files + 1),
                f'Rainfall ({precipitation_column})': rainfall_time_series
            })
             print("\nRainfall Time Series DataFrame:")
             display(rain_ts_df.head())
        else:
            print("\nRainfall time series DataFrame not created.")


    elif len(groundwater_data) == 0 or len(rainfall_data) == 0:
        print("One or both DataFrames are empty. Cannot create time series.")
    else:
        print("Could not determine the number of time periods. Cannot create time series.")

else:
    print("'groundwater_files' or 'rainfall_files' variable not found. Cannot create time series.")
```

```python
from scipy.stats import linregress
import pandas as pd
import numpy as np

# Calculate Groundwater Trend (Change Over Time)
print("Calculating Groundwater Trend...")
groundwater_trend_records = []
# Group groundwater data by location (lat, lon)
groundwater_grouped = groundwater_data.groupby(["lat", "lon"])

for (lat, lon), subset in groundwater_grouped:
    # Ensure there's enough data points for linear regression (at least 2)
    if len(subset) >= 2:
        # Assuming the subset rows are ordered chronologically
        # Create a simple time index based on the number of data points for this location
        time_index = np.arange(len(subset))
        # Ensure there are no NaNs in 'lwe' for linregress (handled by imputation previously, but good practice to check)
        subset_lwe = subset["lwe"].dropna()
        time_index_clean = time_index[subset["lwe"].notna()] # Align time index to non-NaN 'lwe'

        if len(subset_lwe) >= 2:
            # Perform linear regression
            slope, intercept, r, p, se = linregress(time_index_clean, subset_lwe)
            groundwater_trend_records.append({"lat": lat, "lon": lon, "groundwater_trend": slope})
        else:
             # Not enough valid data points for linregress for this location
             groundwater_trend_records.append({"lat": lat, "lon": lon, "groundwater_trend": np.nan})
    else:
        # Not enough data points for linregress for this location
        groundwater_trend_records.append({"lat": lat, "lon": lon, "groundwater_trend": np.nan})


groundwater_trend_df = pd.DataFrame(groundwater_trend_records)
print("Groundwater Trend calculation complete.")
display(groundwater_trend_df.head())

print("\n" + "="*50 + "\n") # Separator


# Calculate Rainfall Trend (Change Over Time)
print("Calculating Rainfall Trend...")
rainfall_trend_records = []
# Group rainfall data by location (lat, lon)
rainfall_grouped = rainfall_data.groupby(["lat", "lon"])

# Use the precipitation column we decided to keep
precipitation_column = 'Precipitation(mm/hr)'

for (lat, lon), subset in rainfall_grouped:
     # Ensure there's enough data points and the precipitation column exists
    if len(subset) >= 2 and precipitation_column in subset.columns:
        # Assuming the subset rows are ordered chronologically
        time_index = np.arange(len(subset))
         # Handle potential NaNs in the precipitation column for linregress
        subset_precip = subset[precipitation_column].dropna()
        time_index_clean = time_index[subset[precipitation_column].notna()] # Align time index

        if len(subset_precip) >= 2:
            # Perform linear regression
            slope, intercept, r, p, se = linregress(time_index_clean, subset_precip)
            rainfall_trend_records.append({"lat": lat, "lon": lon, "rainfall_trend": slope})
        else:
             # Not enough valid data points for linregress for this location
             rainfall_trend_records.append({"lat": lat, "lon": lon, "rainfall_trend": np.nan})
    else:
        # Not enough data points or precipitation column missing for this location
        rainfall_trend_records.append({"lat": lat, "lon": lon, "rainfall_trend": np.nan})


rainfall_trend_df = pd.DataFrame(rainfall_trend_records)
print("Rainfall Trend calculation complete.")
display(rainfall_trend_df.head())
```

```python
import glob
import pandas as pd
import os

# Define the path to the groundwater data directory
groundwater_data_path = '/content/groundwater_data'

# Find all CSV files in the groundwater data directory
groundwater_files = sorted(glob.glob(os.path.join(groundwater_data_path, "*.csv")))

# Check if any groundwater files were found
if not groundwater_files:
    print(f"No CSV files found in {groundwater_data_path}")
else:
    # Read each groundwater CSV file into a DataFrame and store in a list
    groundwater_df_list = []
    for f in groundwater_files:
        try:
            df = pd.read_csv(f)
            groundwater_df_list.append(df)
        except Exception as e:
            print(f"Error reading {f}: {e}")

    # Check if any DataFrames were successfully read
    if groundwater_df_list:
        # Concatenate all groundwater DataFrames in the list into a single DataFrame
        groundwater_data = pd.concat(groundwater_df_list, ignore_index=True)

        # Remove duplicate rows
        initial_groundwater_rows = len(groundwater_data)
        groundwater_data = groundwater_data.drop_duplicates()
        rows_after_groundwater_duplicates = len(groundwater_data)

        print(f"Groundwater data loaded and concatenated.")
        print(f"Rows before dropping duplicates: {initial_groundwater_rows}")
        print(f"Rows after dropping duplicates: {rows_after_groundwater_duplicates}")
        print(f"Duplicate rows removed: {initial_groundwater_rows - rows_after_groundwater_duplicates}")


        # Display the first few rows of the processed groundwater data
        print("\nProcessed Groundwater Data Head (Duplicates Removed):")
        display(groundwater_data.head())

    else:
        print("No groundwater DataFrames were successfully read.")
```
